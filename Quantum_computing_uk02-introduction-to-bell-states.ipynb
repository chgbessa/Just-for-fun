```python
#importing libraryies needed as bibliotecas necessarias

 


from qiskit import QuantumCircuit 
from qiskit import transpile
from qiskit_aer import AerSimulator

from qiskit_ibm_runtime import SamplerV2
from qiskit_ibm_runtime import QiskitRuntimeService

from qiskit.visualization import plot_distribution
```

```python
#Defining backend = Simulator

backend_aer = AerSimulator()
sampler_aer = SamplerV2(backend_aer)
```

First Bell States

```python

#Defining First Bell States



#initializing  quantum circuit



qc1 = QuantumCircuit(2)   #2 qubits

    
qc1.h(0)#apply hadamard gate to the first qubit
qc1.cx(0, 1)#apply cnot gate to both qubits
qc1.measure_all() #qubit measurement
    
print("\n Circuit Bell states 1\n")
qc1.draw("mpl")
```

```python
#Transpiling the circuit, it is needed if we want to send to a real QC

qc1_transpile = transpile(qc1,backend = backend_aer) 

    #Drawing the trasnpiled circuit
    

print("\n Circuit Bell states 1 transpiled\n")
qc1_transpile.draw("mpl")
```

```python
job_aer1 = sampler_aer.run([(qc1_transpile, None, 1000)]) # running the transpiled circuit 1000 times

result1 = job_aer1.result() #Getting the results   

counts1 = result1[0].data.meas.get_counts() #Meassuring results  

print("\n Counts Bell states 1\n")
counts1

    

```

```python
plot_distribution([counts1], legend = ['Aer_simulator--First_Bell_States'])
```
